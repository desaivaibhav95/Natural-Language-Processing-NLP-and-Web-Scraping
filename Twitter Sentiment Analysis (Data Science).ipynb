# Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
```

```python
#Initializing the SentimentIntensityAnalyzer function
sa = SentimentIntensityAnalyzer()

#Putting the polarity scores in the existig data frame as a column
lists = []

for index, rows in data.iterrows():
    score = sa.polarity_scores(rows['Tweets'])
    lists.append(score)
    
Series = pd.Series(lists)
data['polarity'] = Series.values
```
